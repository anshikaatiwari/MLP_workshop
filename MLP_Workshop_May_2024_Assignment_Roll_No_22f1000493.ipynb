## Q.30 Create an adjacency matrix just like the above question, however, this time every `A[i][j]` represents distance between origin `i` and destination `j`. Say there are two flights between `i` and `j`, then the distance is the minimum of the two flights.


```python
df11=dfd[['ORG_AIR','DEST_AIR','DIST']]
df11
```

```python
df_pivot = df11.pivot_table(values='DIST', index='ORG_AIR', columns='DEST_AIR', aggfunc=np.min)

df_pivot.fillna(0, inplace=True)

adjacency_matrix = df_pivot.values
```

```python
adjacency_matrix1=pd.DataFrame(adjacency_matrix,columns=dfkd.columns,index=dfkd.index)
```
